# SEABORN

## INTRODUCTION

* Seaborn is a library for making statistical graphics in Python. It builds on top of matplotlib and integrates closely with pandas data structures.

* Mostly all will prefer to use seaborn than maplotlib Because of it's features.
> Features
>>  * Setting custom color palettes.
>>  * Making attractive statistical plots.
>>  * Easily and flexibly displaying distributions.
>>  * Visualizing information from matrices and DataFrames.

## INSTALLATION

### pip install seaborn and import it

```python
import seaborn as sns
import pandas as pd
import warnings
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# Loading our Data set We Want
auto = pd.read_csv(r"C:\Users\91891\Downloads\archive\Automobile_data.csv")
```

```python
auto.head(6)
```

```python
# LinePlot 1

plt.figure(figsize=(8,10))
sns.lineplot(x = 'wheel-base',y = 'engine-size',data=auto)
plt.show()
```

```python
# Line Plot 2

plt.figure(figsize=(16,6))
sns.set(rc={'axes.facecolor':'#ffe4e1','xtick.labelsize':10, 'ytick.labelsize':10})   #rc is paramater to overwrite values and update parameters
sns.lineplot(data = auto['wheel-base'], linewidth = 1.5, label = 'wheel-base')
sns.lineplot(data = auto['engine-size'], linewidth = 1.5, label = 'engine-size')
plt.xticks(rotation=45)
plt.show()
```

```python
# Line Plot 3

plt.figure(figsize=(14,8))
plt.style.use('seaborn-darkgrid')
sns.lineplot(data = auto, hue = 'fuel-type',  x = 'wheel-base', y = 'engine-size',palette = 'CMRmap', marker = 'o')
plt.legend(bbox_to_anchor=(1.0, 1.0) , shadow=True, fontsize='large')
plt.show()
```

```python
auto['fuel-type']
```

```python
# combining lineplot using relplot
# replot is almost same like subplot in matplotlib
import seaborn as sns
sns.relplot(data = auto , x = 'wheel-base', y = 'engine-size', col = 'fuel-type')
plt.style.use('seaborn-whitegrid')
plt.show()
```

```python
# Scatter Plot

plt.figure(figsize=(6,6))                                                         # gcf denotes  current figure
sns.scatterplot(data = auto , x = 'wheel-base', y = 'engine-size', color = 'red', marker = 'o', hue = 'fuel-type')
plt.gcf().text(.5,.9,"Scatter Plot", fontsize = 40, ha = 'center', va = 'center') # .5 (in x-axis in middle), .9 (y-axis upside)
plt.show()                                                                        # have to take for 0-1.0 (x & y axis)
```

```python
# Joint Plot 

sns.jointplot(x = auto['wheel-base'], y = auto['engine-size'])
plt.show()
```

```python
# Drawing Scatter Plot Using Relplot

sns.relplot(x = auto['wheel-base'], y = auto['engine-size'],hue="fuel-type", kind = 'scatter', palette=["#FF3333" ,"#00CC00"], data = auto)
plt.gcf().text(.5, .99, "Rel Plot", fontsize = 40, color='Black' ,ha='center', va='center')
plt.show()
```

```python

sns.relplot(x = auto['wheel-base'], y = auto['engine-size'], hue = 'fuel-type', data = auto , row = 'body-style')
plt.show()
```

```python
# Bar Plot

plt.figure(figsize=(7,10))
sns.barplot(x = auto['city-mpg'],y = auto['body-style'])
plt.show()
```

```python
# CountPlot in BarPlotplt.figure(figsize=(10,7))
sns.countplot( x = auto['body-style'])   # order=auto['body-style'].value_counts().index (for Ordering It)
plt.show()                       # Used For Counting Purpose
```

```python
# ViolinPlot For Visualising Data
plt.figure(figsize=(10,12))
sns.catplot( x = 'body-style', y = 'city-mpg', kind = 'violin', hue = 'fuel-type' , data = auto)
plt.show()
```

```python
# Strip Plot
# linewidth used for circles width
sns.stripplot( x = 'body-style', y = 'city-mpg', data = auto,linewidth = 0.2, jitter = 0.1) # jitter provide displacement along the x-axis
plt.show()
```

```python
# Dist Plot
#The most convenient way to take a quick look at a univariate distribution in seaborn is the distplot() function.

sns.distplot(x = auto['engine-size'])    # DistPlot = kdeplot + rugplot
warnings.filterwarnings('ignore')
plt.show()
```

```python
# Histogram
plt.figure(figsize=(5,5))
sns.distplot(x = auto['city-mpg'], kde=False, rug=True)
warnings.filterwarnings('ignore')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.distplot(x = auto['city-mpg'], kde=True, rug=True, hist = False)
warnings.filterwarnings('ignore')
plt.show()
```

```python
# Kde Plot

plt.figure(figsize=(10,8))
sns.kdeplot(auto["city-mpg"],auto["engine-size"],shade=True, cmap="Reds")
warnings.filterwarnings('ignore')
plt.show()
```

```python
# Swarm Plot
# Data points won't overlap in Swarm plot
sns.swarmplot(x= "engine-location", y = "engine-size", hue="fuel-type", size = 8 , data = auto)
warnings.filterwarnings('ignore')
plt.show()
```

```python
# Box Plot

sns.boxplot(x=auto["body-style"] ,palette='Set1' , y = auto["engine-size"] , color='black')  # palette is for box color
plt.show()
```

```python
# Pair plot (RelationBetween datasets)

sns.pairplot(auto,hue = 'fuel-type',vars=["wheel-base", "length" , "height" , "width"] )
plt.show()
```

```python
# Regression Plot
# plotting relationship between two variables
plt.figure(figsize=(7,7))
sns.regplot(x=auto["wheel-base"] , y=auto["engine-size"] , marker = '>')
plt.show()
```

```python
# Two Graphs in One

g = sns.JointGrid(x="wheel-base", y="engine-size", data=auto, height = 5)
g = g.plot(sns.regplot, sns.distplot)
```

```python
#  Heat Map

corr = auto.corr()                #co-relation between datasets - corr
plt.figure(figsize=(20,9))
ax = sns.heatmap(corr, annot = True)
plt.yticks(rotation=0)
plt.show()
```
